```python
# importando tabela dos 18 municipios (atualizado para 22)
df_municipios = pd.read_excel(r'dados\municipios_geral.xlsx')
df_municipios.head(10)
```

Tratando a coluna de nomes dos municípios

```python
df_municipios['MUNICIPIO'] = df_municipios['MUNICIPIO'].str.upper()
```

Transformando a coluna de nomes dos municípios em uma lista

```python
# fazendo uma lista de municipios
lista_22_municipios = df_municipios['MUNICIPIO'].tolist()
print(lista_22_municipios)
```

Visualizando lista criada com o nome dos municípios

```python
lista_22_municipios
```

Filtro para mostrar votos que tenham correspondência nos municípios da área de estudo

```python
# condicional para filtrar apenas os resultados que possuam correspondencia nesses municipios
# colunas_candidados = ['NM_MUNICIPIO', 'NM_VOTAVEL','DS_CARGO_PERGUNTA', 'QT_VOTOS', 'NM_PARTIDO', 'SG_PARTIDO']
# array vazio para receber resultados
correspondencias_data = []

# para cada índice por linha presente no df_candidatos
for index, row in df_candidatos.iterrows():
    # se o nome do município estiver na lista de municípios
    if row['NM_MUNICIPIO'] in lista_22_municipios:
        # faça um append no array correspondencias_data
        correspondencias_data.append({
            'MUNICIPIO': row['NM_MUNICIPIO'],
            'NOME_PESSOA': row['NM_VOTAVEL'],
            'CARGO': row['DS_CARGO_PERGUNTA'],
            'VOTOS': row['QT_VOTOS'],
            'NOME_PARTIDO': row['NM_PARTIDO'],
            'SIGLA_PARTIDO': row['SG_PARTIDO']
        })

correspondencias_data
```

Transformando o array de resultados em um dataframe

```python
correspondencias = pd.DataFrame(correspondencias_data)
correspondencias
```

Agrupando dados por votos de cada candidato por município

```python
# agrupando por votos por candidato
corresp_agg = correspondencias.groupby(['MUNICIPIO', 'NOME_PESSOA', 'CARGO'], as_index=False)[['VOTOS']].sum()
corresp_agg
```

Filtrando apenas por deputados federais e salvando em um dataframe

```python
# apenas por deputados federais
filtro_deputados_fed = corresp_agg.loc[corresp_agg['CARGO'].str.contains('Deputado Federal'), :]
filtro_deputados_fed
```

Importando biblioteca beautiful soup para realizar webscrapping

```python
import requests
from bs4 import BeautifulSoup
```

Criando uma função para procurar o usuário do instagram de cada candidato pelo nome encontrado na busca do google

```python


def search_instagram(nome): 
    google_url = f"https://www.google.com/search?q={nome}+instagram" 
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107 Safari/537.36"}
    response = requests.get(google_url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        # encontrar links (a)
        links = soup.find_all('a')

        for link in links:
            href = link.get('href')
            if href and href.startswith("https://www.instagram.com/"):
                return href
    return None

# utilizando eleitos DF
for pessoa in df_deputados_fed_eleitos["nome"]:
    url_instagram = search_instagram(pessoa)
    if url_instagram:
        # adiciona o URL do Instagram na nova coluna de instagram
        df_deputados_fed_eleitos.loc[df_deputados_fed_eleitos["nome"] == pessoa, "Instagram"] = url_instagram
        print(f"Instagram de {pessoa}: {url_instagram}")
    else:
        print(f"Não foi possível encontrar o Instagram de {pessoa}")
```

Pegando lista de deputados estaduais eleitos e os votos de uma página da internet

```python
import re

url = 'https://g1.globo.com/ba/bahia/eleicoes/2022/noticia/2022/10/03/veja-quem-sao-os-63-deputados-estaduais-eleitos-na-bahia.ghtml'


response = requests.get(url)
if response.status_code == 200:
    # parsing do conteúdo HTML
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # encontrando a lista ordenada (ol) que contém os nomes dos candidatos eleitos
    lista_deputados = soup.find('ol')
    
    # verificando se a lista foi encontrada
    if lista_deputados:
        # inicializando listas para armazenar os dados
        nomes = []
        partidos = []
        votos = []
        
        # iterando sobre cada item da lista
        for item in lista_deputados.find_all('li'):
            # extraindo o texto do item
            texto_item = item.text.strip()
            
            # usando regexp para extrair nome, partido e votos
            match = re.match(r'(.*) \((.*?)\): (\d+\.\d+)', texto_item)
            
            # verificando se houve correspondência na regexp
            if match:
                nomes.append(match.group(1))
                partidos.append(match.group(2))
                votos.append(float(match.group(3).replace('.', '')))
            else:
                print("Não foi possível extrair os dados do item:", texto_item)
        
        # df com os dados coletados
        df_deputados_est_eleitos = pd.DataFrame({
            'Nome': nomes,
            'Partido': partidos,
            'Votos': votos
        })
        
        print(df_deputados_est_eleitos)
    else:
        print("Lista de deputados não encontrada.")
else:
    print("Falha ao obter a página. Código de status:", response.status_code)
```

Iterando cada nome do df de deputados estaduais eleito para buscar no google o instagram correspondente por meio da função search_instagram()

```python
for pessoa in df_deputados_est_eleitos["Nome"]:
    url_instagram = search_instagram(pessoa)
    if url_instagram:
        # add a URL do insta na nova coluna instagram
        df_deputados_est_eleitos.loc[df_deputados_est_eleitos["Nome"] == pessoa, "Instagram"] = url_instagram
        print(f"Instagram de {pessoa}: {url_instagram}")
    else:
        print(f"Não foi possível encontrar o Instagram de {pessoa}")
```
